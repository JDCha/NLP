```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import somoclu
%matplotlib inline
```

```python
df = pd.read_csv('oneday_vector.csv', encoding="utf-8")
game_id = df.pop('game_id')

s = pd.DataFrame.as_matrix(df)
game_id = pd.DataFrame.as_matrix(game_id)
```

```python
index = []
for i in range(len(s)):
    if np.count_nonzero(s[i]) < 200:
        index.append(i)

s = np.delete(s, index, 0)
game_id = np.delete(game_id, index, 0)
print(len(s), len(game_id))
```

```python
s
```

```python
n_rows, n_columns = 200, 320
som = somoclu.Somoclu(n_columns, n_rows, data=s)
%time som.train()
```

```python
som.cluster()
som.view_umatrix(bestmatches=True, figsize=(20, 20))
```

```python
som.view_umatrix(bestmatches=True, figsize=(20,20), labels=game_id)
```

```python
def get_id_terms(id):
    term_list = []
    value = np.array(view.loc[view['game_id'] == id])[0]
    for index, i in enumerate(value):
        if not i == 0:
            term_list.append(view.columns[index])
    print(term_list)
```

```python
print("yellow"), get_id_terms(9707)
```
