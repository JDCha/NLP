```python
data = pd.read_csv('./cluster_label.csv')
data.head()
```

```python
data.columns
```

```python
data_0 = data.loc[data['real_label']=='C0']
data_1 = data.loc[data['real_label']=='C1']
data_2 = data.loc[data['real_label']=='C2']
data_3 = data.loc[data['real_label']=='C3']
```

```python
def filter_strange(object):
    try:
        return int(object)
    except:
        return 0
```

```python
data_1['ind_i']=data_1['ind_i'].apply(filter_strange)
```

```python
# data_1['ship']=data_1['ship'].apply(filter_strange)
# data_1['jail']=data_1['jail'].apply(filter_strange)

fig, ax = plt.subplots(figsize=(8,8))

ax.scatter(data_0['ship'],data_0['jail'],color='g',alpha=.4,label='C0')
ax.scatter(data_1['ship'],data_1['jail'],color='r',alpha=.4,label='C1')
ax.scatter(data_2['ship'],data_2['jail'],color='b',alpha=.4,label='C2')
ax.scatter(data_3['ship'],data_3['jail'],color='y',alpha=.4,label='C3')
## Make a legend at upper right
plt.legend(loc='upper right')
## X, Y label name
plt.xlabel('ship')
plt.ylabel('jail')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.scatter(data_0['a5'],data_0['a7'],color='g',alpha=.4,label='C0')
ax.scatter(data_1['a5'],data_1['a7'],color='r',alpha=.4,label='C1')
ax.scatter(data_2['a5'],data_2['a7'],color='b',alpha=.4,label='C2')
ax.scatter(data_3['a5'],data_3['a7'],color='y',alpha=.4,label='C3')
## Make a legend at upper right
plt.legend(loc='upper right')
## X, Y label name
plt.xlabel('Topic 5 - party')
plt.ylabel('Topic 7 - assult')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

data_0['pvpdi']=data_0['pvpdi'].apply(filter_strange)
data_1['pvpdi']=data_1['pvpdi'].apply(filter_strange)
data_2['pvpdi']=data_2['pvpdi'].apply(filter_strange)
data_3['pvpdi']=data_3['pvpdi'].apply(filter_strange)

graph = [data_0['pvpdi'],data_1['pvpdi'],data_2['pvpdi'],data_3['pvpdi']]


bp = ax.boxplot(graph, patch_artist = True)

bp['boxes'][0].set(color='#000000', linewidth=1)
bp['boxes'][0].set( facecolor = '#67f776' )
bp['boxes'][1].set(color='#000000', linewidth=1)
bp['boxes'][1].set( facecolor = '#f92c33' )
bp['boxes'][2].set(color='#000000', linewidth=1)
bp['boxes'][2].set( facecolor = '#3131ed' )
bp['boxes'][3].set(color='#000000', linewidth=1)
bp['boxes'][3].set( facecolor = '#fffc4f' )

bp['medians'][0].set(color='#000000', linewidth=2)
bp['medians'][1].set(color='#000000', linewidth=2)
bp['medians'][2].set(color='#000000', linewidth=2)
bp['medians'][3].set(color='#000000', linewidth=2)
    # get position data for median line
ax.grid(alpha=.4,linewidth=.5,linestyle='--')
ax.set_xticklabels(['C0', 'C1','C2','C3'])
plt.title('complete quest')

plt.show()
```
